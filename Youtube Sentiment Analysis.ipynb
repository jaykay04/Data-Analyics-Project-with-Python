# Youtube Comments
## Sentiment Analysis 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
videos_data1=pd.read_csv('UScomments.csv',error_bad_lines=False)
```

```python
videos_data1.head()
```

```python
videos_data1.shape
```

```python
#To perform Sentiments Analysis, we need to install the textblob library.
#We can install it from here or command prompt.

!pip install textblob
```

```python
from textblob import TextBlob
```

```python
TextBlob('Logan Paul it\'s yo big day ‚ÄºÔ∏è‚ÄºÔ∏è‚ÄºÔ∏è	').sentiment.polarity
```

```python
polarity_result=[]
for i in videos_data1['comment_text']:
    try:
        polarity_result.append(TextBlob(i).sentiment.polarity)
    except:
        polarity_result.append(0)
```

```python
videos_data1['result']=polarity_result
```

```python
videos_data1.head(20)
```

# Exploratory Data Analysis(EDA) for Positive & Negative Sentiments
## WordCloud representation of sentiments

```python
#Wordclouds are basically used for text data
positive_comments=videos_data1[videos_data1['result']==1]
positive_comments.shape[0]
```

```python
positive_comments.head()
```

```python
#First install wordcloud library
!pip install wordcloud
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
total_comments=' '.join(positive_comments['comment_text'])
len(total_comments)
```

```python
cloud_visuals=WordCloud(width=1000,height=500,stopwords=set(STOPWORDS)).generate(total_comments)
plt.figure(figsize=(15,5))
plt.imshow(cloud_visuals)
plt.axis('off')
```

```python
negative_comments=videos_data1[videos_data1['result']==-1]
negative_comments.head()
```

```python
total_negcomments=' '.join(negative_comments['comment_text'])
len(total_negcomments)
```

```python
cloud_visuals=WordCloud(width=1000,height=500,stopwords=set(STOPWORDS)).generate(total_negcomments)
plt.figure(figsize=(15,5))
plt.imshow(cloud_visuals)
plt.axis('off')
```
